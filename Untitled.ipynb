```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/Skyserver_SQL2_27_2018 6_51_39 PM.csv")
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
gal = df[df["class"] == "GALAXY"]
gal.shape
```

```python
gal = gal[["u", "g", "r", "i", "z", "redshift"]].dropna()
gal.head()
```

```python
plt.scatter(gal["redshift"], gal["r"], s=2)
plt.xlabel("Redshift (distance)")
plt.ylabel("r-band brightness")
plt.title("Galaxy brightness vs redshift")
plt.gca().invert_yaxis()  # brighter = lower number
plt.show()
```

```python
gal["u_g"] = gal["u"] - gal["g"]
gal["g_r"] = gal["g"] - gal["r"]
gal.head()
```

```python
plt.scatter(gal["u_g"], gal["g_r"], s=2)
plt.xlabel("u - g (color)")
plt.ylabel("g - r (color)")
plt.title("Galaxy color-color diagram")
plt.show()
```

```python
gal["redshift"].corr(gal["u_g"])
```

```python
gal["redshift"].corr(gal["g_r"])
```

```python
plt.scatter(gal["g_r"], gal["r"], s=2)
plt.gca().invert_yaxis()
plt.xlabel("g - r (color)")
plt.ylabel("r-band brightness")
plt.title("Galaxy color-magnitude diagram")
plt.show()
```

```python
from sklearn.cluster import KMeans

X = gal[["g_r", "r"]].values

kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
gal["pop"] = kmeans.labels_
```

```python
plt.scatter(gal["g_r"], gal["r"], c=gal["pop"], s=2)
plt.gca().invert_yaxis()
plt.xlabel("g - r (color)")
plt.ylabel("r-band brightness")
plt.title("Galaxy populations (red vs blue)")
plt.show()
```
