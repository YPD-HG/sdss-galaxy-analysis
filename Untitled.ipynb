```python
from sklearn.cluster import KMeans

X = gal[["g_r", "r"]].values

kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
gal["pop"] = kmeans.labels_
```

```python
# The universe had more actively star forming galaxies in the past, and galaxies became redder and more passive over time.
```

```python
# Now we do, Quantitative scaling relations.
# We will quantify:

# How fast galaxy color changes with redshift

# How star formation declines with time
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = gal["redshift"].values.reshape(-1, 1)
y = gal["u_r"].values

model = LinearRegression().fit(X, y)

slope = model.coef_[0]
intercept = model.intercept_

slope, intercept

# This gives us, how fast galaxies become redder with time
```

```python
# Above result tells us, as redshift increases by 1, the average u  r color changes by about 0.33 magnitudes.

# Because lower u  r means bluer, and galaxies at higher redshift are bluer, this tells us:
# galaxies become significantly bluer as we go back in time, which means star formation was higher in the past.
```

```python
for p in gal["pop"].unique():
    subset = gal[gal["pop"] == p]
    X = subset["redshift"].values.reshape(-1,1)
    y = subset["u_r"].values
    model = LinearRegression().fit(X,y)
    print("Population", p, "slope:", model.coef_[0])
```

```python
# One populations color changes slowly with redshift.
# The other populations color changes much more strongly with redshift.
```

```python
# The red, passive galaxies evolve very differently from the blue, star-forming galaxies.
# One population quenches rapidly. The other evolves slowly.
```

```python
# Now, adding error bars to your u  r vs redshift evolution plot.
```

```python
stats = gal.groupby("z_bin")["u_r"].agg(["mean", "std", "count"])
stats["stderr"] = stats["std"] / (stats["count"] ** 0.5)
stats
```

```python
# Implying:

# Nearby galaxies have u  r  2.2  0.01

# Distant galaxies have u  r  0.9  0.3
```

```python
import numpy as np

z_centers = stats.index.map(lambda x: (x.left + x.right)/2)

plt.errorbar(z_centers, stats["mean"], yerr=stats["stderr"], fmt="o")
plt.xlabel("Redshift")
plt.ylabel("Mean u - r color")
plt.title("Galaxy color evolution with cosmic time")
plt.show()
```
