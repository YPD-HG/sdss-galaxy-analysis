```python
# Now we will see how fast galaxy color (and therefore star formation) changes with cosmic time.
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Use the binned statistics you already computed
z = stats.index.map(lambda x: (x.left + x.right) / 2).values.reshape(-1,1)
color = stats["mean"].values
err = stats["stderr"].values

# Remove bins with NaNs
mask = ~np.isnan(color) & ~np.isnan(err)
z = z[mask]
color = color[mask]
err = err[mask]

# Weighted linear regression
weights = 1 / err**2
model = LinearRegression()
model.fit(z, color, sample_weight=weights)

slope = model.coef_[0]
intercept = model.intercept_

slope, intercept

# This gives us the rate at which galaxies stop forming stars as the universe ages.
```

```python
# This is showing as the universe ages (redshift decreases), galaxies become redder at a measurable rate.
```

```python
# The mean u  r color evolves as u  r  2.10 + 0.78 z, implying a strong, statistically measured decline in star formation with cosmic time.
```

```python
# Now we will overlay our fitted model on the binned data
```

```python
# Numeric redshift centers
z_centers = stats.index.map(lambda x: (x.left + x.right)/2).values

# Data
color = stats["mean"].values
err = stats["stderr"].values

# Remove NaNs
mask = ~np.isnan(color) & ~np.isnan(err)
z_centers = z_centers[mask]
color = color[mask]
err = err[mask]

# Plot SDSS data with error bars
plt.errorbar(
    z_centers,
    color,
    yerr=err,
    fmt="o",
    label="SDSS data",
    capsize=4
)

# Plot fitted evolution law
z_fit = np.linspace(z_centers.min(), z_centers.max(), 100).reshape(-1,1)
color_fit = model.predict(z_fit)

plt.plot(z_fit, color_fit, label="Weighted linear fit", linewidth=2)

plt.xlabel("Redshift")
plt.ylabel("Mean u  r color")
plt.title("Measured cosmic evolution of galaxy color")
plt.legend()
plt.show()
```

```python
# Graph description:

# Each blue point is the average color of thousands of galaxies in a slice of cosmic time.

# The vertical bars are the uncertainty in that average.

# The orange line is a weighted regression that trusts precise bins more and noisy bins less.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("data/Skyserver_SQL2_27_2018 6_51_39 PM.csv")

gal = df[df["class"] == "GALAXY"].copy()
gal = gal[["u","g","r","i","z","redshift"]].dropna()

print("Total galaxies:", len(gal))
gal.head()
```

```python
# Calculating absolute magnitude below
```

```python
# Cosmology constants
c = 299792.0   # km/s
H0 = 70.0     # km/s/Mpc

z = gal["redshift"].values

# Luminosity distance in Mpc (good approximation for z < 0.3)
DL = (c / H0) * z * (1 + z)

# Avoid log(0)
DL[DL == 0] = np.nan

# Absolute r-band magnitude
gal["M_r"] = gal["r"] - 5 * np.log10(DL) - 25

gal[["redshift","r","M_r"]].head()
```

```python
# Note that here M_r is true galaxy brightness
```

```python
plt.scatter(gal["redshift"], gal["M_r"], s=3)
plt.gca().invert_yaxis()   # brighter = up
plt.xlabel("Redshift")
plt.ylabel("Absolute magnitude M_r")
plt.title("Which galaxies SDSS detects at different redshifts")
plt.show()
```

```python
# Curved edge on the bottom is SDSSs detection limit.

# It means:
# At high redshift, SDSS can only see very bright galaxies.
# At low redshift, it sees bright and faint ones.

# That is the bias we must remove.
```

```python
# Now we find a brightness cut that SDSS can see at all redshifts.
```

```python
# Look only at distant galaxies
highz = gal[gal["redshift"] > 0.2]

# Faintest galaxy SDSS still detects there
M_limit = highz["M_r"].max()

print("Volume-limited magnitude cut:", M_limit)
```

```python
# Meaning at high redshift, SDSS can only see galaxies brighter than about M_r = 20.87.
```

```python
# Now we will build the volume-limited sample
```

```python
vol = gal[gal["M_r"] <= M_limit].copy()

print("Original galaxies:", len(gal))
print("Volume-limited galaxies:", len(vol))
```

```python
# We now have 2,896 galaxies that SDSS would detect at every redshift.
# That means we can now compare the same type of galaxies through time
```

```python
# Now we recompute galaxy color for this clean sample.
```

```python
vol["u_r"] = vol["u"] - vol["r"]
vol["u_r"].describe()
```

```python
# Split this clean sample into redshift bins.
```

```python
bins = [0.0, 0.05, 0.1, 0.2, 0.4]
vol["z_bin"] = pd.cut(vol["redshift"], bins)

stats = vol.groupby("z_bin")["u_r"].agg(["mean","std","count"])
stats["stderr"] = stats["std"] / np.sqrt(stats["count"])

stats
```

```python
# At low redshift
# u  r  2.53

# At higher redshift
# u  r  1.46

# Even after removing all selection bias, galaxies were still much bluer in the past.
```

```python
import pandas as pd
import numpy as np

# Load SDSS data
df = pd.read_csv("data/Skyserver_SQL2_27_2018 6_51_39 PM.csv")

# Keep only galaxies
df = df[df["class"] == "GALAXY"].copy()

# Remove bad values
df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna(subset=["u", "g", "r", "z"])

# Compute color
df["u_r"] = df["u"] - df["r"]

# Compute absolute magnitude (needed for volume limit)
# SDSS gives luminosity distance via redshift; this is a standard approximation
# M = m - 5 log10(D_L) - 25
c = 3e5  # km/s
H0 = 70
df["D_Mpc"] = (c * df["z"]) / H0
df["M_r"] = df["r"] - 5 * np.log10(df["D_Mpc"]) - 25

df.head()
```

```python
# Now volume limited sample
```

```python
# Define redshift limit of the survey
z_max = df["z"].max()

# Find the faintest galaxy we could still see at z_max
M_limit = np.percentile(df["M_r"], 90)

# Build volume limited sample
vol = df[df["M_r"] < M_limit].copy()

print("Full sample:", len(df))
print("Volume limited sample:", len(vol))
print("Absolute magnitude limit:", M_limit)
```

```python
# We removed the ~10 percent of galaxies
```

```python
# Use median color in volume limited sample to split populations
color_cut = vol["u_r"].median()

# Split into blue and red
blue = vol[vol["u_r"] < color_cut]
red  = vol[vol["u_r"] >= color_cut]

print("Color cut (u-r):", color_cut)
print("Blue galaxies:", len(blue))
print("Red galaxies:", len(red))
```

```python
# Now we will measure color evolution separately for blue and red galaxies, but still in the volume limited sample.
```

```python
# Define redshift bins
bins = [0.0, 0.05, 0.1, 0.2, 0.4]

def binned_stats(df):
    df = df.copy()
    df["z_bin"] = pd.cut(df["redshift"], bins)
    stats = df.groupby("z_bin")["u_r"].agg(["mean","std","count"])
    stats["stderr"] = stats["std"] / np.sqrt(stats["count"])
    return stats

blue_stats = binned_stats(blue)
red_stats  = binned_stats(red)

blue_stats, red_stats
```

```python
# Extract bin centers
z_blue = blue_stats.index.categories.mid.astype(float)
u_blue = blue_stats["mean"].values
err_blue = blue_stats["stderr"].values

z_red = red_stats.index.categories.mid.astype(float)
u_red = red_stats["mean"].values
err_red = red_stats["stderr"].values
```

```python
import numpy as np

Nboot = 5000

slopes_blue = []

for i in range(Nboot):
    # resample the bins with replacement
    idx = np.random.choice(len(z_blue), len(z_blue), replace=True)
    
    zb = z_blue[idx]
    ub = u_blue[idx]
    eb = err_blue[idx]
    
    w = 1 / eb**2
    
    # weighted linear fit
    b, a = np.polyfit(zb, ub, 1, w=w)
    slopes_blue.append(b)

slopes_blue = np.array(slopes_blue)

b_blue = slopes_blue.mean()
sig_blue = slopes_blue.std()

print("Blue galaxies:")
print("slope =", b_blue)
print("sigma =", sig_blue)
print("significance =", abs(b_blue)/sig_blue)
```

```python
# The above result tells us
# We do not have statistically strong evidence that blue galaxy color is changing with redshift in the volume limited sample.
# For blue galaxies alone, SDSS volume limited sample does not have enough statistics to strongly constrain the slope.
```

```python
# Now we do the red galaxies
```

```python
slopes_red = []

for i in range(Nboot):
    idx = np.random.choice(len(z_red), len(z_red), replace=True)
    
    zr = z_red[idx]
    ur = u_red[idx]
    er = err_red[idx]
    
    w = 1 / er**2
    
    b, a = np.polyfit(zr, ur, 1, w=w)
    slopes_red.append(b)

slopes_red = np.array(slopes_red)

b_red = slopes_red.mean()
sig_red = slopes_red.std()

print("Red galaxies:")
print("slope =", b_red)
print("sigma =", sig_red)
print("significance =", abs(b_red)/sig_red)
```

```python
# Similar to blue, there is no statistically significant color evolution for red galaxies in the volume limited SDSS sample.
```
